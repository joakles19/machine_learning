# Titanic survival predictor

___

```python
import pandas as pd
from ml_algorithms import RandomForest
```

```python
titanic_df = pd.read_csv('data/titanic_data.csv')
titanic_df.head()
```

```python
data = titanic_df[['Sex', 'Age', 'Fare', 'Family_size', 'Survived']]
data.head()
```

```python
test_split = int(len(data) * 0.8)
train_data = data[:test_split]
test_data = data[test_split:]
```

```python
forest = RandomForest(40)
forest.fit_data(train_data.values)
```

```python
predictions = forest.predict(test_data.values)
test_data['Prediction'] = predictions
```

```python
correct = test_data[test_data['Survived'] == test_data['Prediction']]

print(len(correct) / len(test_data))
```
